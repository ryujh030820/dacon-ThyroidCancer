# 데이터 분석

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터 로드 (가정)
df = pd.read_csv('train.csv')

# 인종별 갑상선암 발생률 계산
race_cancer = df.groupby('Race')['Cancer'].mean().reset_index()
race_cancer['Cancer'] = race_cancer['Cancer'] * 100  # 퍼센트로 변환

# 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x='Race', y='Cancer', data=race_cancer)
plt.title('Thyroid Cancer Prevalence by Race (%)')
plt.ylabel('Cancer Prevalence (%)')
plt.xlabel('Race')
plt.show()
```

```python
# Pandas 출력 설정 조정
pd.set_option('display.max_columns', None)  # 모든 열 표시
pd.set_option('display.max_rows', None)     # 모든 행 표시
pd.set_option('display.width', None)        # 너비 제한 해제
pd.set_option('display.max_colwidth', None) # 열 너비 제한 해제

# 인종별 환자 수
race_counts = df['Race'].value_counts()
print("인종별 환자 수:\n", race_counts)

# 인종별 나이, 결절 크기, TSH/T4/T3 분포
race_stats = df.groupby('Race')[['Age', 'Nodule_Size', 'TSH_Result', 'T4_Result', 'T3_Result']].describe()
print("\n인종별 주요 변수 통계:\n", race_stats)
```

```python
# 인종별 위험 요인과 갑상선암 상관관계
risk_factors = ['Family_Background', 'Radiation_History', 'Iodine_Deficiency', 'Smoke', 'Weight_Risk', 'Diabetes']
for factor in risk_factors:
    cross_tab = pd.crosstab(df['Race'], df[factor], normalize='index')
    print(f"\n인종별 {factor} 비율:\n", cross_tab)
    
    # 갑상선암과의 연관성
    cancer_cross = pd.crosstab([df['Race'], df[factor]], df['Cancer'], normalize='index')
    print(f"\n{factor}와 갑상선암 연관성:\n", cancer_cross)
```

```python
# 인종 및 성별별 갑상선암 발생률
gender_race_cancer = df.groupby(['Race', 'Gender'])['Cancer'].mean().reset_index()
gender_race_cancer['Cancer'] = gender_race_cancer['Cancer'] * 100

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='Race', y='Cancer', hue='Gender', data=gender_race_cancer)
plt.title('Thyroid Cancer Prevalence by Race and Gender (%)')
plt.ylabel('Cancer Prevalence (%)')
plt.xlabel('Race')
plt.show()
```

```python
# 인종별 생리학적 변수와 갑상선암 상관관계
bio_vars = ['Nodule_Size', 'TSH_Result', 'T4_Result', 'T3_Result']
for var in bio_vars:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Race', y=var, hue='Cancer', data=df)
    plt.title(f'{var} by Race and Cancer Status')
    plt.show()
```

```python
# 범주형 변수를 수치형으로 인코딩
df_encoded = df.copy()
df_encoded['Gender'] = df_encoded['Gender'].map({'M': 0, 'F': 1})
df_encoded['Race'] = df_encoded['Race'].map({'ASN': 0, 'CAU': 1, 'AFR': 2, 'HSP': 3, 'MDE': 4})
df_encoded['Family_Background'] = df_encoded['Family_Background'].map({'Positive': 1, 'Negative': 0})
df_encoded['Radiation_History'] = df_encoded['Radiation_History'].map({'Exposed': 1, 'Unexposed': 0})
df_encoded['Iodine_Deficiency'] = df_encoded['Iodine_Deficiency'].map({'Deficient': 1, 'Sufficient': 0})
df_encoded['Smoke'] = df_encoded['Smoke'].map({'Smoker': 1, 'Non-Smoker': 0})
df_encoded['Weight_Risk'] = df_encoded['Weight_Risk'].map({'Obese': 1, 'Not Obese': 0})
df_encoded['Diabetes'] = df_encoded['Diabetes'].map({'Yes': 1, 'No': 0})

# Country 열 인코딩
label_encoder = LabelEncoder()
df_encoded['Country'] = label_encoder.fit_transform(df_encoded['Country'])

# 상관분석에 사용할 feature 선택 (Country 포함, ID 제외)
features = ['Age', 'Gender', 'Country', 'Race', 'Family_Background', 'Radiation_History', 
            'Iodine_Deficiency', 'Smoke', 'Weight_Risk', 'Diabetes', 
            'Nodule_Size', 'TSH_Result', 'T4_Result', 'T3_Result', 'Cancer']

# 상관계수 행렬 계산
correlation_matrix = df_encoded[features].corr(method='pearson')

# 상관계수 히트맵 시각화
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Features (Including Country)')
plt.show()

# 상관계수 행렬 출력 (텍스트로 확인)
print("\nCorrelation Matrix:\n")
print(correlation_matrix)
```
